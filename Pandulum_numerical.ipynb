```python
import numpy as np
import matplotlib.pyplot as plt 

t0 = 0
x0 = 1
xdot0 = 0
tf = 100
n = 100001
g = 9.81 # the gravity constant [m/s^2]
l = 1    # the length of the rod of the pendulum [m]
b = 0.2  #viscous friction coefficient [Nms/rad]
m =  1   #the mass of the bob of the pendulum [kg]
```

```python
dt = (tf - x0) / (n - 1)
t = np.linspace(t0,tf, n)
x = np.zeros ([ n ])
xdot = np.zeros ([ n ])
xddot = np.zeros ([ n ])
```

```python
x[0] = x0
xdot[0] = xdot0
xddot[0] = 0
for i in range (1 , n) :

    xddot[i] = -(g/l)* np.sin(x[i-1]) - ((b/(m*(l**2)))*xdot[i-1])
    xdot[i] = xdot[i-1] +  dt * xddot[i]
    x[i] =  x[i-1] + (dt * (xdot[i]))
    
# for i in range(n):
#     print (x[i], y[i])
```

```python
plt.plot(t,  x)
plt.xlabel('time')
plt.ylabel('x [rad]')
# plt.title('Approximate S ol u ti o n with Forward Euler Method')
plt.show ()
```

```python
plt.plot(t,  xdot)
plt.xlabel('time')
plt.ylabel('xdot [m]')
# plt.title('Approximate S ol u ti o n with Forward Euler Method')
plt.show ()
```
